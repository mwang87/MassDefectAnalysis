```python
import statsmodels.api as sm

component_df = clusterinfo_df[clusterinfo_df["componentindex"] == "2"]

rlm_model = sm.RLM(component_df["defect"], component_df["precursor mass"], M=sm.robust.norms.HuberT())
rlm_results = rlm_model.fit()
print(rlm_results.params)
```

```python
#Linear Regression

from sklearn import linear_model
clt = linear_model.LinearRegression()
component_df = clusterinfo_df[clusterinfo_df["componentindex"] == "2"]

clt.fit(pd.DataFrame(component_df["precursor mass"]), pd.DataFrame(component_df["defect"]))

print("SCORE", clt.score(pd.DataFrame(component_df["precursor mass"]), pd.DataFrame(component_df["defect"])))
print(clt.coef_, clt.intercept_)

yhat = clt.predict(pd.DataFrame(component_df["precursor mass"]))

plot_df = pd.DataFrame()
plot_df["precursor mass"] = component_df["precursor mass"]
plot_df["yhat"] = yhat

ax = sns.scatterplot(x="precursor mass", y="yhat", palette=sns.color_palette("Set1", 17), \
                     data=plot_df)

ax = sns.scatterplot(x="precursor mass", y="defect", \
                     data=clusterinfo_df[clusterinfo_df["componentindex"] == "2"])

plt.show()
```

```python
from sklearn import linear_model

def determine_component_regressions(component_df):
    #Determine the number of unique integer masses
    unique_integer_masses = list(set([int(mass) for mass in list(component_df["precursor mass"])]))
    
    #print("UNIQUE MASSES", len(unique_integer_masses))
    
    clt = linear_model.RANSACRegressor(linear_model.LinearRegression())
    clt.fit(pd.DataFrame(component_df["precursor mass"]), pd.DataFrame(component_df["defect"]))
    score = clt.score(pd.DataFrame(component_df["precursor mass"]), pd.DataFrame(component_df["defect"]))

    return len(unique_integer_masses), score, clt.estimator_.coef_, clt.estimator_.intercept_
```

```python
#Reading the data
import requests

TASK_ID = "c95481f0c53d42e78a61bf899e9f9adb"
try:
    clusterinfo_df = pd.DataFrame(requests.get("https://gnps.ucsd.edu/ProteoSAFe/result_json.jsp?task=%s&view=view_all_clusters_withID_beta" % (TASK_ID)).json()["blockData"])
except:
    clusterinfo_df = pd.DataFrame(requests.get("https://gnps.ucsd.edu/ProteoSAFe/result_json.jsp?task=%s&view=view_all_clusters_withID" % (TASK_ID)).json()["blockData"])

clusterinfo_df["precursor mass"] = clusterinfo_df["precursor mass"].astype(float)
#clusterinfo_df["defect"] = clusterinfo_df["precursor mass"].apply(lambda num : float(num) - int(float(num))) 
clusterinfo_df["defect"] = clusterinfo_df["precursor mass"].apply(lambda num : int((0.998883) * (float(num))) - \
                                                                  (0.998883) * (float(num))) 

unique_components = list(clusterinfo_df["componentindex"].unique())
unique_components.sort()

for component in unique_components:
    if component == "-1":
        continue
    filtered_df = clusterinfo_df[clusterinfo_df["componentindex"] == component]

    try:
        number_uniq, score, slope, intercept = determine_component_regressions(filtered_df)
        
        if number_uniq < 3 or score < 0.9:
            continue
        print(component, number_uniq, score, slope, intercept)
    except:
        continue
```
